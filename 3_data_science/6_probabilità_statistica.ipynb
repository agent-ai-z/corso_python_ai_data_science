```python

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
    
```

## **1. Fondamenti di Probabilità**
La **probabilità** è la misura della possibilità che un evento accada.

- **Spazio campionario (S)**: L'insieme di tutti i possibili risultati.
- **Evento (E)**: Un sottoinsieme dello spazio campionario.
- **Probabilità di un evento (P(E))**: La misura della possibilità di E, con \( 0 \leq P(E) \leq 1 \).

**Formula della probabilità classica:**
$$P(E) = \frac{|E|}{|S|}$$

dove $|E|$ è il numero di risultati favorevoli e $|S|$ il numero totale di risultati.

### **Esempio: Lancio di un dado**

```python

# Spazio campionario (dado a 6 facce)
spazio_campionario = [1, 2, 3, 4, 5, 6]

# Evento: ottenere un numero pari
evento = [2, 4, 6]

# Calcolo della probabilità
probabilità = len(evento) / len(spazio_campionario)
print(f"La probabilità di ottenere un numero pari è: {probabilità:.2f}")
    
```

## **2. Distribuzioni di Probabilità**
Le **distribuzioni di probabilità** descrivono il comportamento di una variabile casuale.

### **Distribuzione Uniforme**
- **Proprietà**: Ogni valore ha la stessa probabilità di essere scelto.
- **Esempio**: Lancio di un dado.

```python

# Generazione di numeri casuali da una distribuzione uniforme
uniform_data = np.random.uniform(0, 10, 1000)

# Creazione dell'istogramma
plt.figure(figsize=(6, 4))
sns.histplot(uniform_data, bins=30, kde=True, color="blue")
plt.title("Distribuzione Uniforme")
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.show()
    
```

### **Distribuzione Normale (Gaussiana)**
- **Proprietà**: La maggior parte dei valori è concentrata attorno alla media.
- **Esempio**: Altezza delle persone, errori di misura.

```python

# Generazione di dati casuali con distribuzione normale
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Creazione dell'istogramma
plt.figure(figsize=(6, 4))
sns.histplot(normal_data, bins=30, kde=True, color="green")
plt.title("Distribuzione Normale")
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.show()
    
```

### **Distribuzione Binomiale**
- **Proprietà**: Modella il numero di successi in \( n \) prove indipendenti.
- **Esempio**: Numero di teste in 10 lanci di una moneta.

```python

# Generazione di dati con distribuzione binomiale
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)

# Creazione dell'istogramma
plt.figure(figsize=(6, 4))
sns.histplot(binomial_data, bins=10, kde=False, color="red")
plt.title("Distribuzione Binomiale")
plt.xlabel("Numero di Successi")
plt.ylabel("Frequenza")
plt.show()
    
```

## **3. Statistica Descrittiva**
La **statistica descrittiva** aiuta a riassumere e descrivere i dati.

- **Media**: La media aritmetica dei dati.
- **Mediana**: Il valore centrale dei dati ordinati.
- **Moda**: Il valore più frequente.

### **Calcolo delle Statistiche Descrittive con Pandas**

```python

# Creazione di un DataFrame con dati casuali
np.random.seed(42)
df = pd.DataFrame({"Età": np.random.randint(18, 60, 100)})

# Calcolo delle statistiche
media = df["Età"].mean()
mediana = df["Età"].median()
moda = df["Età"].mode()[0]

# Stampa dei risultati
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
    
```

### **Misure di Variabilità**
Le **misure di dispersione** ci dicono quanto i dati sono dispersi attorno alla media.

- **Varianza** $sigma^2$: Media dei quadrati degli scarti dalla media.
- **Deviazione standard** $sigma$: Radice quadrata della varianza.

```python

# Calcolo della varianza e deviazione standard
varianza = df["Età"].var()
deviazione_standard = df["Età"].std()

# Stampa dei risultati
print(f"Varianza: {varianza:.2f}")
print(f"Deviazione Standard: {deviazione_standard:.2f}")
    
```

## **4. Raccolta delle Misurazioni Statistiche**
La raccolta delle misure statistiche aiuta a comprendere le caratteristiche dei dati.

- **Deviazione Standard**: Misura la dispersione dei dati rispetto alla media.
- **Range**: Differenza tra il valore massimo e minimo.
- **Asimmetria (Skewness)**: Indica se i dati sono simmetrici o sbilanciati.
- **Curtosi**: Misura la distribuzione dei dati rispetto a una curva normale.

```python

# Creazione di un DataFrame con dati casuali
np.random.seed(42)
df = pd.DataFrame({"Valori": np.random.normal(50, 10, 1000)})

# Calcolo delle misure statistiche
dev_standard = df["Valori"].std()
range_val = df["Valori"].max() - df["Valori"].min()
asimmetria = df["Valori"].skew()
curtosi = df["Valori"].kurtosis()

# Stampa dei risultati
print(f"Deviazione Standard: {dev_standard:.2f}")
print(f"Range: {range_val:.2f}")
print(f"Asimmetria (Skewness): {asimmetria:.2f}")
print(f"Curtosi: {curtosi:.2f}")
    
```

### **Visualizzazione della Distribuzione dei Dati**
Utilizziamo un **istogramma** con la curva di densità per osservare la distribuzione dei dati.

```python

# Creazione dell'istogramma
plt.figure(figsize=(6, 4))
sns.histplot(df["Valori"], bins=30, kde=True, color="blue")
plt.title("Distribuzione dei Dati")
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.show()
    
```

#### **Correlazione tra Due Variabili**
La **correlazione** misura la relazione tra due variabili.

- **Coefficiente di Pearson**: Indica la relazione lineare tra due variabili.
- **Coefficiente di Spearman**: Valuta relazioni monotone non necessariamente lineari.
- **Valori tra -1 e 1**:
  - **1** → Relazione perfettamente positiva
  - **0** → Nessuna relazione
  - **-1** → Relazione perfettamente negativa

```python

# Creazione di dati correlati
np.random.seed(42)
x = np.random.rand(100) * 10  # Variabile indipendente
y = 2.5 * x + np.random.normal(0, 2, 100)  # Variabile dipendente con rumore

# Creazione del DataFrame
df_corr = pd.DataFrame({"X": x, "Y": y})

# Calcolo della correlazione di Pearson e Spearman
pearson_corr = df_corr.corr(method="pearson").iloc[0, 1]
spearman_corr = df_corr.corr(method="spearman").iloc[0, 1]

# Stampa dei risultati
print(f"Coefficiente di Pearson: {pearson_corr:.2f}")
print(f"Coefficiente di Spearman: {spearman_corr:.2f}")
    
```

##### **Visualizzazione della Relazione tra Due Variabili**

```python

# Creazione del grafico a dispersione
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df_corr["X"], y=df_corr["Y"], color="red")
plt.title("Grafico a Dispersione tra X e Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
    
```

### **Esempio: Test di Student**
Verifichiamo se la media di due campioni è significativamente diversa.

```python

# Creazione di due campioni di dati casuali
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)

# Test t di Student
t_stat, p_value = stats.ttest_ind(group1, group2)

# Risultati
print(f"Statistiche t: {t_stat:.4f}, p-value: {p_value:.4f}")

# Interpretazione
if p_value < 0.05:
    print("Le medie dei due gruppi sono significativamente diverse.")
else:
    print("Non ci sono differenze significative tra le medie dei due gruppi.")
    
```
