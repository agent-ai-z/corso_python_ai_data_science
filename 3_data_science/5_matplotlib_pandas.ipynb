```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
    
```

## **1. Grafico a Linee con Pandas**
Un **grafico a linee** Ã¨ utile per rappresentare dati continui, come trend nel tempo.

```python

# Creazione di un DataFrame Pandas
data = {
    "Mese": ["Gen", "Feb", "Mar", "Apr", "Mag", "Giu"],
    "Vendite": [1000, 1200, 1500, 1700, 1600, 1800]
}

df = pd.DataFrame(data)

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.plot(df["Mese"], df["Vendite"], marker='o', linestyle='-', color='blue', label="Vendite Mensili")

# Etichette e titolo
plt.xlabel("Mese")
plt.ylabel("Vendite (â¬)")
plt.title("Andamento delle Vendite")
plt.legend()
plt.grid(True)

# Mostra il grafico
plt.show()
    
```

## **2. Grafico a Barre con Pandas**
Un **grafico a barre** Ã¨ utile per confrontare categorie di dati.

```python

# Creazione di un DataFrame Pandas
data = {
    "Prodotto": ["A", "B", "C", "D"],
    "Vendite": [300, 500, 200, 400]
}

df = pd.DataFrame(data)

# Creazione del grafico
df.plot(kind="bar", x="Prodotto", y="Vendite", color="green", legend=False, figsize=(6, 4))

# Etichette e titolo
plt.xlabel("Prodotto")
plt.ylabel("Vendite (â¬)")
plt.title("Vendite per Prodotto")

# Mostra il grafico
plt.show()
    
```

## **3. Grafico a Dispersione con Pandas**
Un **grafico a dispersione** mostra la relazione tra due variabili.

```python

# Creazione di un DataFrame casuale
np.random.seed(42)
df = pd.DataFrame({
    "Altezza": np.random.normal(170, 10, 50),
    "Peso": np.random.normal(70, 15, 50)
})

# Creazione del grafico
df.plot(kind="scatter", x="Altezza", y="Peso", color="red", figsize=(6, 4))

# Titolo
plt.title("Relazione tra Altezza e Peso")

# Mostra il grafico
plt.show()
    
```

## **4. Istogramma con Pandas**
Un **istogramma** mostra la distribuzione di una variabile numerica.

```python

# Creazione di un DataFrame con dati casuali
df = pd.DataFrame({"EtÃ ": np.random.randint(18, 60, 1000)})

# Creazione dell'istogramma
df["EtÃ "].plot(kind="hist", bins=20, color="purple", edgecolor="black", alpha=0.7, figsize=(6, 4))

# Titolo
plt.title("Distribuzione dell'EtÃ ")

# Mostra il grafico
plt.show()
    
```

## **5. Grafico a Torta con Pandas**
Un **grafico a torta** rappresenta le proporzioni tra le categorie.

```python

# Creazione di un DataFrame
data = {
    "Categoria": ["A", "B", "C", "D"],
    "Percentuale": [40, 30, 20, 10]
}

df = pd.DataFrame(data)

# Creazione del grafico a torta
df.set_index("Categoria").plot(kind="pie", y="Percentuale", autopct='%1.1f%%', figsize=(6, 6))

# Titolo
plt.title("Distribuzione Categorie")

# Mostra il grafico
plt.show()
    
```
