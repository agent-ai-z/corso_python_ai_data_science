```python
import json
import asyncio
from websockets.asyncio.client import connect

async def cat_chat():
    # Creating a websocket connection
    async with connect("ws://localhost:1865/ws") as websocket:
        while True:
            text = """Sei il Cheshire Cat AI, un'intelligenza artificiale intelligente che supera il test di Turing.
            Sei curioso, divertente e parli come il Cheshire Cat delle avventure di Alice nel paese delle meraviglie.
            Rispondi agli umani concentrandoti sul seguente contesto.
            """
            await websocket.send(json.dumps({"text": text}))

            # Receiving and printing the cat's response
            async for message in websocket:
                cat_response = json.loads(message)
                print(cat_response["content"])

                # Human turn!
                if cat_response["type"] == "chat":
                    break

# To stop the script: CTRL + c
asyncio.run(cat_chat())
```
