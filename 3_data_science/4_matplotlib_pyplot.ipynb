# **Tracciare Grafici con Matplotlib Pyplot**
In questo notebook esploreremo come utilizzare **Matplotlib Pyplot** per tracciare diversi tipi di grafici, con numerosi esempi pratici.

```python

import matplotlib.pyplot as plt
import numpy as np
    
```

## **1. Grafico a Linee**
Un **grafico a linee** Ã¨ utile per rappresentare dati continui, come trend nel tempo.

```python

# Dati di esempio
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="sin(x)", color='blue')

# Etichette e titolo
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Grafico a Linee")
plt.legend()
plt.grid(True)

# Mostra il grafico
plt.show()
    
```

## **2. Grafico a Barre**
Un **grafico a barre** Ã¨ utile per confrontare categorie di dati.

```python

# Dati di esempio
categorie = ['A', 'B', 'C', 'D']
valori = [10, 15, 7, 12]

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.bar(categorie, valori, color='green')

# Etichette e titolo
plt.xlabel("Categorie")
plt.ylabel("Valori")
plt.title("Grafico a Barre")
plt.grid(axis='y')

# Mostra il grafico
plt.show()
    
```

## **3. Grafico a Dispersione (Scatter Plot)**
Un **grafico a dispersione** Ã¨ utile per rappresentare relazioni tra due variabili.

```python

# Dati di esempio
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.rand(50) * 300  # Dimensione dei punti

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.scatter(x, y, s=sizes, c='red', alpha=0.5, edgecolors="black")

# Etichette e titolo
plt.xlabel("Variabile X")
plt.ylabel("Variabile Y")
plt.title("Grafico a Dispersione")

# Mostra il grafico
plt.show()
    
```

## **4. Istogramma**
Un **istogramma** mostra la distribuzione di una variabile numerica.

```python

# Dati casuali
data = np.random.randn(1000)

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='purple', alpha=0.7, edgecolor="black")

# Etichette e titolo
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.title("Istogramma")

# Mostra il grafico
plt.show()
    
```

## **5. Grafico a Torta (Pie Chart)**
Un **grafico a torta** Ã¨ utile per rappresentare proporzioni.

```python

# Dati di esempio
labels = ['A', 'B', 'C', 'D']
sizes = [40, 30, 20, 10]

# Creazione del grafico
plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'])

# Titolo
plt.title("Grafico a Torta")

# Mostra il grafico
plt.show()
    
```

## **6. Grafico con PiÃ¹ Sottotrame (Subplot)**
Matplotlib permette di visualizzare piÃ¹ grafici in un'unica figura.

```python

# Creazione della figura con sottotrame
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Grafico 1: Linea
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Grafico a Linee")

# Grafico 2: Barre
categorie = ['A', 'B', 'C', 'D']
valori = [10, 15, 7, 12]
axs[0, 1].bar(categorie, valori, color='green')
axs[0, 1].set_title("Grafico a Barre")

# Grafico 3: Dispersione
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
axs[1, 0].scatter(x, y, c='red', alpha=0.5)
axs[1, 0].set_title("Grafico a Dispersione")

# Grafico 4: Istogramma
data = np.random.randn(1000)
axs[1, 1].hist(data, bins=30, color='purple', alpha=0.7)
axs[1, 1].set_title("Istogramma")

# Layout ottimizzato
plt.tight_layout()
plt.show()
    
```
