# **Problema del Commesso Viaggiatore (TSP)**
In questo notebook esploreremo il **Problema del Commesso Viaggiatore (TSP)**, la sua definizione, metodi di risoluzione e implementazione con **NetworkX**.

## **1. Definizione del Problema**
Il **Problema del Commesso Viaggiatore (TSP - Traveling Salesman Problem)** consiste nel trovare il **percorso più breve** che visita un insieme di città esattamente una volta e torna alla città di partenza, minimizzando la distanza totale percorsa.

### **Formula Matematica**
Data una lista di **n** città e le distanze tra ciascuna coppia di città, il problema può essere scritto come:
\[ \min \sum_{i=1}^{n} d(c_i, c_{i+1}) \]
dove:
- \( c_i \) è la città **i-esima**.
- \( d(c_i, c_{i+1}) \) è la distanza tra le città \( c_i \) e \( c_{i+1} \).
- \( c_n \) ritorna alla città iniziale.

## **2. Complessità del Problema**
Il TSP è un problema **NP-difficile**, il che significa che **non esiste un algoritmo polinomiale noto** per risolverlo esattamente.

- Per **n città**, il numero totale di possibili percorsi è **(n-1)! / 2** (per grafi non orientati).
- Per **n > 20**, il numero di soluzioni diventa **esplosivo** e richiede metodi approssimati.

## **3. Algoritmi per Risolvere il TSP**
Esistono diversi approcci per risolvere il problema, sia esatti che approssimati:

### **3.1. Algoritmi Esatti (Per Grafi Piccoli)**
- **Brute Force**: Prova tutte le permutazioni possibili (**O(n!)**).
- **Programmazione Dinamica (Held-Karp)**: Riduce il costo a **O(n² 2ⁿ)**.
- **Branch and Bound**: Riduce lo spazio di ricerca scartando soluzioni non ottimali.

### **3.2. Algoritmi Euristici (Per Grafi Grandi)**
- **Nearest Neighbor (NN)**: Sceglie sempre la città più vicina (**O(n²)**, non garantisce l'ottimo).
- **Algoritmo di Christofides**: Approssimazione con fattore **3/2**.

### **3.3. Algoritmi Metaeuristici (Ottimizzazione)**
- **Algoritmi Genetici (GA)**: Evoluzione simulata della soluzione.
- **Simulated Annealing (SA)**: Simula un processo termodinamico per evitare minimi locali.
- **Ant Colony Optimization (ACO)**: Simula il comportamento delle formiche.

```python

import networkx as nx
import matplotlib.pyplot as plt
import random
    
```

## **4. Creazione di un Grafo Pesato per il TSP**

```python

# Creazione di un grafo completo con 6 città
G = nx.complete_graph(6)
pos = nx.spring_layout(G)

# Assegniamo pesi casuali alle distanze tra città
for (u, v) in G.edges():
    G[u][v]['weight'] = random.randint(1, 100)

# Disegniamo il grafo con i pesi sugli archi
labels = nx.get_edge_attributes(G, 'weight')
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=12)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
    
```

## **5. Risoluzione Approssimata del TSP con NetworkX**

```python

# Algoritmo euristico per il TSP
tsp_cycle = nx.approximation.traveling_salesman_problem(G, cycle=True)

print(f"Percorso trovato: {tsp_cycle}")
    
```
