### Alternativa: Usare una Variabile d'Ambiente

```python

import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(api_key=os.getenv('OPENAI_API_KEY')) # Imposta la chiave come variabile d'ambiente per sicurezza
```

### Crea una risposta simile a quella umana a un prompt

```python
PROMPT = "Cos'è il machine learning?"

completion = client.chat.completions.create(
    model=OPENAI_MODEL,
    messages=[
        {
            "role": "system", 
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": PROMPT
        }
    ]
)

print(completion.choices[0].message.content)
```

### Generazione di un immagine

```python
import requests
from IPython.display import Image, display

try:
    
    # PROMPT = "Un cucciolo di foca tra i ghiacchi dell'antartico"
    PROMPT = "Crea un disegno di un cucciolo di foca con lo stile di un pittore impressionista"

    response = client.images.generate(
        prompt=PROMPT,
        size="512x512"
    )

    # Download the image
    image_url = response.data[0].url
    img_data = requests.get(image_url).content

    # Save the image
    with open('openai_img.png', 'wb') as handler:
        handler.write(img_data)

    print("Image downloaded and saved as openai_img.png")
    display(Image(filename='openai_img.png'))
except Exception as e:
    print(f"An error occurred: {e}")
    print("Please check if your project has access to the required model.")
```
